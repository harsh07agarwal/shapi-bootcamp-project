```python
import hashlib
print("WELCOME TO A HASH GENERATOR PROGRAM")
string=input("input the string:")
while(True):
  print("\nMAIN MANU")
  print("1. MD5 Hash")
  print("2. SHA256 Hash")
  print("3. SHA1 Hash")
  print("4. SHA512 Hash")
  print("5. Blake2s Hash")
  print("6. Exit")
  choice=int(input("Enter the choice:"))

  if choice==1:
    result=hashlib.md5(string.encode())
    print("the byte equivalent of md5 hash is: ",result.digest())
    print("the hexadecimal equivalent of md5 hash is: ",result.hexdigest(),"\n")
  
  elif choice==2:
    res=hashlib.sha256(string.encode())
    print("the byte equivalent of sha256 hash is: ",res.digest())
    print("the hexadecimal equivalent of sha256 hash is: ",res.hexdigest(),"\n")

  elif choice==3:
    n=hashlib.sha1(string.encode())
    print("the byte equivalent of sha1 hash is: ",n.digest(),)
    print("the hexadecimal equivalent of sha1 hash is: ",n.hexdigest(),"\n")

  elif choice==4:
    enc=hashlib.sha512(string.encode())
    print("the byte equivalent of sha512 hash is: ",enc.digest())
    print("the hexadecimal equivalent of sha512 hash is: ",enc.hexdigest(),"\n")

  elif choice==5:
    enc1=hashlib.blake2s(string.encode())
    print("the byte equivalent of blake2s hash is: ",enc1.digest())
    print("the hexadecimal equivalent of blake2s hash is: ",enc1.hexdigest(),"\n")

  elif choice==6:
    break

  else:  
    print("Oops! Incorrect Choice.")  

  
```
